```python
# 성별에따른 월급 평균 산출
gender_income = df.dropna(subset=['income']) # income 컬럼을 기준으로 na가 있으면 관측치(레코드, 행)를 제거
print(gender_income.shape)

gender_income = gender_income.groupby(['gender'], as_index=False) # groupby에 사용된 컬럼을 일반 컬럼으로 처리
print(gender_income)

gender_income = gender_income.agg(income_mean = ('income', 'mean'))
print(gender_income)
#    gender  income_mean
# 0  female   190.617382 여자
# 1    male   365.106375 남자
```

```python
# 결합
gender_income = df.dropna(subset=['income']).groupby(['gender'], as_index=False).agg(income_mean = ('income', 'mean'))
print(gender_income)
#    gender  income_mean
# 0  female   190.617382 여자
# 1    male   365.106375 남자
```

```python
# 평균 급여를 백분율로 나타낼 것.
print(type(gender_income))
print(gender_income['income_mean'])    # 1차원 배열
print(type(gender_income['income_mean']))
print(gender_income['income_mean'][0]) # 1차원 배열의 요소
print(gender_income['income_mean'][1])
```

```python
female = gender_income['income_mean'][0]
male = gender_income['income_mean'][1]

print(female / male)       # 비율 나누기
print((female / male)*100) # 백분율
print(f'여성의 평균 급여는 남성 100% 기준 {round((female / male)*100, 1)}%임') # 소수 첫째자리까지 반올림
```

#### IT 직군의 성별 평균 급여
- 0222	컴퓨터 시스템 및 소프트웨어 전문가
- 0223	데이터 및 네트워크 관련 전문가
- 0224	정보 시스템 및 웹 운영자

```python
print(df['code_job'].dtypes)
print(df['code_job'].head(5))
```

```python
# 필요한 변수만 선정
it_df = df[['gender', 'income', 'code_job']].dropna(subset=['income']) # income에 값이 없으면 레코드 삭제
it_df = it_df.query('code_job in [222, 223, 224]')
it_df.head(5)
```

```python
print(it_df.shape)
```

```python
print(it_df['gender'].value_counts())
```

```python
it_income = it_df.groupby(['gender'], as_index=False).agg(income_mean = ('income', 'mean'))
it_income
```

```python
female = it_income['income_mean'][0]
male = it_income['income_mean'][1]

print(female / male)       # 비율 나누기
print((female / male)*100) # 백분율
print(f'여성의 평균 급여는 남성 100% 기준 {round((female / male)*100, 1)}%임') # 소수 첫째자리까지 반올림
# 2019년 남성 100% 기준 여성은 85.7%였음.
```

```python
여성의 평균 급여는 남성 100% 기준 77.2%로 파악되었다.
여성의 급여가 낮은 이유는 여러가지 이유중에 난이도 높은 SW 개발 분야에 근무하는 비율이 적은것으로 볼 수 있고,
특히 여성의 표본이 너무 적어 이 통계는 대표성을 갖기 어려운것으로 보인다.
너무 적은 표보은 특정값에 편중 될수도 있음으로 더 많은 표본을 수집할 필요가 있다.
이 통계는 인정하기 어렵다.
```
